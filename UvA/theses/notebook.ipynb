```python
url = 'https://scripties.uba.uva.nl/search?f1-faculteit-nl=Faculteit%20der%20Maatschappij-%20en%20Gedragswetenschappen%20(FMG);startDoc='
url_list = []
for i in range (1, 7373, 10):
    url_list.append(url + str(i))

url_list[:10]
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

id_list_FMG = []
for url in tqdm(url_list):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for link in soup.find_all('a', class_='c-item__link'):
        id_list_FMG.append(link.get('href'))
```

```python
url = 'https://scripties.uba.uva.nl/search?f1-faculteit-nl=Faculteit%20der%20Geesteswetenschappen%20(FGw);startDoc='
url_list_H = []
for i in range (1, 3283, 10):
    url_list_H.append(url + str(i))

url_list_H[:10]
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

id_list_H = []
for url in tqdm(url_list_H):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for link in soup.find_all('a', class_='c-item__link'):
        id_list_H.append(link.get('href'))
```

```python
len(id_list_H)
```

```python
theses_FMG = []
for thesis in tqdm(id_list_FMG):
    response= requests.get("https://scripties.uba.uva.nl/"+thesis)

    soup = BeautifulSoup(response.text, 'html.parser')

    try:
        title = soup.find('th', string='Titel').find_next_sibling('td').text.strip()
    except:
        title = None
    try:
        author = soup.find('th', string='Auteur(s)').find_next_sibling('td').text.strip()
    except:
        author = None
    
    try:
        supervisor = soup.find('th', string='Supervisor(s)').find_next_sibling('td').text.strip()
    except:
        supervisor = None
    
    try:
        year = soup.find('th', string='Jaar').find_next_sibling('td').text.strip()
    except:
        year = None
    
    try:
        faculty = soup.find('th', string='Faculteit').find_next_sibling('td').text.strip()
    except:
        faculty = None
    
    try:
        program = soup.find('th', string='Opleiding').find_next_sibling('td').text.strip()
    except:
        program = None
    
    try:
        language = soup.find('th', string='Taal').find_next_sibling('td').text.strip()
    except:
        language = None

    try:
        abstract = soup.find('th', string='Abstract').find_next_sibling('td').text.strip()
    except:
        abstract = None
    
    #make dict
    thesis_dict = {'title': title, 'author': author, 'supervisor': supervisor, 'year': year, 'faculty': faculty, 'program': program, 'language':language, 'abstract': abstract}
    theses_FMG.append(thesis_dict)

```

```python
theses_H = []
for thesis in tqdm(id_list_H):
    response= requests.get("https://scripties.uba.uva.nl/"+thesis)

    soup = BeautifulSoup(response.text, 'html.parser')

    try:
        title = soup.find('th', string='Titel').find_next_sibling('td').text.strip()
    except:
        title = None
    try:
        author = soup.find('th', string='Auteur(s)').find_next_sibling('td').text.strip()
    except:
        author = None
    
    try:
        supervisor = soup.find('th', string='Supervisor(s)').find_next_sibling('td').text.strip()
    except:
        supervisor = None
    
    try:
        year = soup.find('th', string='Jaar').find_next_sibling('td').text.strip()
    except:
        year = None
    
    try:
        faculty = soup.find('th', string='Faculteit').find_next_sibling('td').text.strip()
    except:
        faculty = None
    
    try:
        program = soup.find('th', string='Opleiding').find_next_sibling('td').text.strip()
    except:
        program = None
    
    try:
        language = soup.find('th', string='Taal').find_next_sibling('td').text.strip()
    except:
        language = None

    try:
        abstract = soup.find('th', string='Abstract').find_next_sibling('td').text.strip()
    except:
        abstract = None
    
    #make dict
    thesis_dict = {'title': title, 'author': author, 'supervisor': supervisor, 'year': year, 'faculty': faculty, 'program': program, 'language':language, 'abstract': abstract}
    theses_H.append(thesis_dict)

```

```python
#make dataframes
import pandas as pd
df_FMG = pd.DataFrame(theses_FMG)
df_H = pd.DataFrame(theses_H)

#save to csv
df_FMG.to_csv('FMG.csv')
df_H.to_csv('Humanities.csv')
```

```python
import pandas as pd
from nltk.corpus import stopwords

df = pd.read_csv('UvA_theses.csv')

eng = df[df['lang'] == 'en']
doc_list = eng['abstract'].tolist()
```

```python
from bertopic import BERTopic

topic_model = BERTopic(language="english", verbose=True)
topics, probs = topic_model.fit_transform(doc_list)

topic_model.get_topic_freq().head(10)
```

```python
#visualize
import matplotlib.pyplot as plt
topic_model.visualize_topics()
```

```python
#hierarchy
topic_model.visualize_hierarchy()
```

```python
#fit to df
eng['Topic'] = topics
eng
```

```python
#topics over time with bertopic

#first, convert the year to datetime
eng['year'] = pd.to_datetime(eng['year'], format='%Y')


topics_over_time = topic_model.topics_over_time(doc_list, eng['year'].tolist())
topic_model.visualize_topics_over_time(topics_over_time, top_n_topics=10)
```
